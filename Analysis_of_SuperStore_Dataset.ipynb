# **Building** **a** **Machine** **Learning** **model** **to** **Predict** **the** **Profit**

```python
data.head()
```

```python
df1=data.drop(columns=["Order ID","Country","City","State","Product Name","Sub-Category"])
df1.head()
```

```python
df1.shape
```

```python
df1["Ship Mode"].value_counts()
```

```python
df1["Segment"].value_counts()
```

```python

df1["Category"].value_counts()
```

```python
df1["Region"].value_counts()
```

```python
#converting Categorical variable into numerical value
df1.replace({"Ship Mode":{"Standard Class":0,"Second Class":1,"First Class":2,"Same Day":3},"Segment":{"Consumer":0,"Corporate":1,"Home Office":2},
                 "Category":{"Office Supplies":0,"Furniture":1,"Technology":2},"Region":{"West":0,"East":1,"Central":2,"South":3}},inplace=True)
```

```python
df1.head()
```

```python
x=df1.iloc[:,:-1]
print(x)
```

```python
y=df1.iloc[:,-1]
print(y)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
num_vars=["Sales","Quantity"]
x[num_vars]=scaler.fit_transform(x[num_vars])
print(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=100)
```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
#Creating Machine Learning Model
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor(random_state=0,n_estimators=117)
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
print("R2 Score=",r2_score(y_pred,y_test)*100)
```

```python
fig = plt.figure()
plt.scatter(y_test,y_pred)
fig.suptitle('y_test vs y_pred', fontsize=20)
plt.xlabel('y_test', fontsize=18)
plt.ylabel('Y_pred', fontsize=16)
```
